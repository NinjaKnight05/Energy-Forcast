```python
df['Datetime'] = pd.to_datetime(df['Datetime'])
```

```python
df.head()
```

```python
df.info()
```

```python
df.sort_values("Datetime",inplace = True)
```

```python
df.duplicated().sum()
```

```python
df.duplicated("Datetime").sum()
```

```python
df.drop_duplicates("Datetime",inplace = True)
```

```python
df.duplicated("Datetime").sum()
```

```python
full_range = pd.date_range(start=df['Datetime'].min(), end=df['Datetime'].max(), freq='h')
missing = full_range.difference(df['Datetime'])
print(len(missing))
```

```python
df.set_index("Datetime",inplace = True)
```

```python
df = df.resample('h').mean()
```

```python
df = df.interpolate()
```

```python
df.index
```

```python
df.shape
```

```python
Yearly = df.resample('YE').mean()
px.line(Yearly, y='AEP_MW', title = 'Yearly Average Enerfy Consumption')
```

```python
monthly = df.resample('ME').mean()
px.line(monthly,y='AEP_MW', title='Montly Average Energy Consumption')
```

```python
df['hour'] = df.index.hour
hourly = df.groupby('hour')['AEP_MW'].mean()
px.bar(hourly, title='Average Consumption by Hour of Day')
```

```python
df['hour'] = df.index.hour
df['dayofweak'] = df.index.dayofweek
df['month'] = df.index.month
df['year'] = df.index.year
df['quarter'] = df.index.quarter
```

```python
df['dayofyear'] = df.index.dayofyear
df['weekofyear'] = df.index.isocalendar().week.astype(int)
df['lag_1'] = df['AEP_MW'].shift(1)
df['lag_24'] = df['AEP_MW'].shift(24)
df['lag_168'] = df['AEP_MW'].shift(168)
df = df.dropna()
```

```python
df
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x= df[['hour', 'dayofweak', 'month', 'year', 'quarter', 'dayofyear', 'weekofyear', 'lag_1', 'lag_24', 'lag_168']]
y = df['AEP_MW']

xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,shuffle=False,)
```

```python
# %pip install xgboost
```

```python
from xgboost import XGBRegressor
model = XGBRegressor()
```

```python
model.fit(xtrain,ytrain)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

yPred = model.predict(xtest)
```

```python
mae = mean_absolute_error(ytest,yPred)
```

```python
rmse = np.sqrt(mean_squared_error(ytest,yPred))
```

```python
print(f'MAE: {mae:.2f}')
print(f'RMSE: {rmse:.2f}')
```

```python
print(r2_score(ytest,yPred))
```

```python
print(r2_score(ytrain, model.predict(xtrain)))
print(r2_score(ytest, model.predict(xtest)))
```
